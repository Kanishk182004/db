```
from diffusers import StableDiffusionPipeline
from transformers import CLIPTokenizer # Import CLIPTokenizer
import torch

# Load the model and tokenizer
model_id = "CompVis/stable-diffusion-v1-4"
tokenizer = CLIPTokenizer.from_pretrained(model_id, subfolder="tokenizer") # Load tokenizer from subfolder
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")  # Assuming you have a GPU

def generate_ecom_image(prompt, num_inference_steps=50, guidance_scale=7.5):
  """
  Generates an image of an e-commerce product based on the given text prompt.

  Args:
    prompt: The text description of the product.
    num_inference_steps: The number of denoising steps.
    guidance_scale: The strength of the guidance scale.

  Returns:
    The generated image.
  """
  image = pipe(prompt, num_inference_steps=num_inference_steps, guidance_scale=guidance_scale).images[0]
  return image

# Example Usage
product_description = "A black sports shoe with metalic finish and nike symbol."
generated_image = generate_ecom_image(product_description)

# Save the generated image
generated_image.save("generated_product_image.png")
```
